# Histogramas Frecuencias Absolutas
No referimos al numero de veces que un valor o categorias especificas aparece en un conjunto de datos.

```python
# Generamos datos aleatorios de alturas
alturas = np.random.normal(170, 10, 500) # media de 170 cm y desviacion estandar de 10 cm
```

```python
# Creamos el histograma
plt.hist(alturas, bins=20, color='skyblue', edgecolor='black') # bins = 20 intervalos
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Alturas')
```

# Histograma de freciencias Relativas
Proporcion o porcentaje que representa la frecuencia absoluta con respecto al total de observaciones

```python
plt.hist(alturas, bins=20, color='skyblue', edgecolor='black', density=True)
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia relativa')
plt.title('Distribuci√≥n de Alturas')
```

# Series de tiempo

```python
import pandas as pd
import numpy as np

# ------------------------------
# GENERAR DATOS DE PRECIPITACI√ìN
# ------------------------------

# Crear rango de fechas mensuales desde 2000 hasta 2020
dates = pd.date_range(start="2000-01-01", end="2020-12-31", freq="MS")

# Generar valores aleatorios de precipitaci√≥n entre 20 y 200 mm
np.random.seed(0)  # para reproducibilidad
precip = np.random.uniform(20, 200, len(dates))

# Crear DataFrame
df = pd.DataFrame({
    "Fecha": dates,
    "Precipitacion_mm": np.round(precip, 2)
})

# Mostrar primeras filas
print("Primeras filas del DataFrame:")
print(df.head())

# ------------------------------
# GUARDAR CSV
# ------------------------------

csv_filename = "precipitaciones_2000_2020.csv"
df.to_csv(csv_filename, index=False)

print("\nArchivo CSV generado correctamente:", csv_filename)

# ------------------------------
# DESCARGAR CSV EN COLAB
# ------------------------------
from google.colab import files
files.download(csv_filename)
```

```python
# Creamos grafico de lineas
# Convertir la columna Fecha a tipo fecha
df["Fecha"] = pd.to_datetime(df["Fecha"])

# Extraer el mes y el a√±o en nuevas columnas
df["A√±o"] = df["Fecha"].dt.year
df["Mes"] = df["Fecha"].dt.month

# --- Filtrar solo AGOSTO (mes 8) ---
df_agosto = df[df["Mes"] == 8]
```

```python
# Crear grafico de lineas
plt.plot(df_agosto["A√±o"], df_agosto["Precipitacion_mm"], marker="o")
plt.xlabel("A√±o")
plt.ylabel("Precipitaci√≥n en Agosto (mm)")
plt.title("Precipitaci√≥n en agosto a lo largo de los a√±os")
plt.grid(True)
plt.show()
```

# Seaborn

```python
import pandas as pd
import numpy as np

# Rango de a√±os
years = range(2001, 2021)  # podes cambiar el rango que quieras

# Meses en ingl√©s
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# DataFrame vac√≠o
df = pd.DataFrame({"Year": years})

# Fijamos semilla (opcional) para reproducibilidad
np.random.seed(0)

# Generar datos de precipitaci√≥n aleatorios por mes
for month in months:
    df[month] = np.round(np.random.uniform(0, 800, len(years)), 3)

# Mostrar primeras filas
df.head()
```

```python
# Crear grafico de barras
sns.barplot(x="Year", y="Aug", data=df)
plt.xlabel("A√±o")
plt.ylabel("Precipitaci√≥n en Agosto (mm)")
plt.title("Precipitaci√≥n en Agosto a lo largo de los a√±os")
plt.xticks(rotation=90)
plt.show()
```

```python
# Personalizacion del grafico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x="Year", y="Aug", data=df, ci='sd', palette="viridis", order=df.sort_values('Aug', ascending=False)['Year'])
plt.xlabel("A√±o", fontsize=14)
plt.ylabel("Precipitaci√≥n en Agosto (mm)", fontsize=14)
plt.title("Precipitaci√≥n en Agosto a lo largo de los a√±os", fontsize=16)
plt.xticks(rotation=90)
plt.xlabel("A√±o", fontsize=14)
plt.show()
```

```python
 # Crear una figura con 2 filas y 2 columnas de subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
# Graficar en cada subplot
sns.histplot(data=df, x='variable1', ax=axs[0, 0])
sns.boxplot(data=df, x='variable2', ax=axs[0, 1])
sns.scatterplot(data=df, x='variable3', y='variable4', ax=axs[1, 0])
sns.lineplot(data=df, x='variable5', y='variable6', ax=axs[1, 1])
plt.tight_layout()
plt.show()
```
