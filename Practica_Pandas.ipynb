```python
import pandas as pd
import numpy as np

# Crear un data frame con valores ausentes
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, 7, 8],
    'C': [9, 10, 11, np.nan]})

# Detectar valores ausentes
print(df.isnull())

```

```python
df_dropna = df.dropna()
print(df_dropna)
```

```python
df_fillna = df.fillna(0)
print(df_fillna)
```

```python
df_fillna_mean =df.fillna(df.mean())
print(df_fillna_mean)
```
