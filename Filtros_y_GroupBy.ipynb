```python
import pandas as pd
# URL del archivo CSV
url = 'https://raw.githubusercontent.com/brunobobadilla06/Bruno-CienciaDeDatos/refs/heads/main/stocks_data.csv'
df_stockdata = pd.read_csv(url)

df_stockdata['formatted_date'] = pd.to_datetime(df_stockdata['formatted_date'])
condicion_fecha = df_stockdata['formatted_date'] >= pd.Timestamp('2016-03-01')
df_desde_marzo = df_stockdata[condicion_fecha]


# Codicion multiple
df_combo1 = df_stockdata[(df_stockdata['GOOG'] > 700) & (df_stockdata['MSFT'] < 1000)]
print(df_combo1)
```

```python
condicion_marzo = df_stockdata['formatted_date'] >= '2016-03-01'
condicion_amazon_alto = df_stockdata['AMZN'] > 2500
df_combo2 = df_stockdata[condicion_marzo | condicion_amazon_alto]
print(df_combo2)
```

```python
# Agrupacion con groupby()
# Necesitamos extraer componentes de la fecha (Ao, Mes, Trimestre) para agrupar
df_stockdata['year'] = df_stockdata['formatted_date'].dt.year
df_stockdata['month'] = df_stockdata['formatted_date'].dt.month
df_stockdata['quarter'] = df_stockdata['formatted_date'].dt.quarter

df_stockdata.head(10)
```

```python
ticker_cols = ['MCD', 'SBUX', 'GOOG', 'AMZN', 'MSFT', 'JPM', 'BAC', 'C', 'MAR', 'HLT', 'RCL', 'V', 'MA', 'PYPL']

# CALCULAR EL PRECIO PROMEDIO ANUAL PARA CADA TICKETS
# Agrupamos por ao, y luego calculamos la media de todas las columnas numericas
promedio_anual = df_stockdata.groupby('year')[ticker_cols].mean()
promedio_anual.head()
```

```python
# Calcular el precio maximo por trimestre para cada ticket
max_trimestre_ticket = df_stockdata.groupby(['year', 'quarter'])[ticker_cols].max()
max_trimestre_ticket.head()
```

```python

import pandas as pd
# URL del archivo CSV
url = 'https://raw.githubusercontent.com/brunobobadilla06/Bruno-CienciaDeDatos/refs/heads/main/stocks_data.csv'
df_stockdata = pd.read_csv(url)

df_stockdata['formatted_date'] = pd.to_datetime(df_stockdata['formatted_date'])
condicion_fecha = df_stockdata['formatted_date'] >= pd.Timestamp('2016-03-01')
df_desde_marzo = df_stockdata[condicion_fecha]
df_stockdata['year'] = df_stockdata['formatted_date'].dt.year

# Contar cuantos registros (meses) hay por ao
registros_por_ao = df_stockdata.groupby('year').size()
registros_por_ao.head()
```

```python
# Agregacion multiples con .agg()
# Aplicamos .agg() despues de agrupar por tiempo, especificando que hacer con cada columna de tickets(o las mismas operaciones para varias)
agregaciones_seleccion = {
    'GOOG': ['min', 'max', 'mean', 'std'],
    'AMZN': ['min', 'max', 'mean', 'std'],
    'MSFT': ['min', 'max', 'mean', 'std']
}
Prom_des__std_por_ao = df_stockdata.groupby('year').agg(agregaciones_seleccion)
Prom_des__std_por_ao.head()
```

```python
# Calcular el promedio anual para todos los tickets, pero el maximo anual para AMZN
#DEFINIMOS AGREGACIONES PARA TODAS LAS COLUMNAS DE TICKETS Y UNA ESPECIFICA PARA AMZN
ticker_cols = ['MCD', 'SBUX', 'GOOG', 'AMZN', 'MSFT', 'JPM', 'BAC', 'C', 'MAR', 'HLT', 'RCL', 'V', 'MA', 'PYPL']
agg_dict = {ticket: 'mean' for ticket in ticker_cols}

agg_dict['AMZN'] = ['mean', 'max']

resumen_combinado = df_stockdata.groupby('year').agg(agg_dict)
resumen_combinado.head()
```

```python
# Aplicando funciones a los datos como apply, map
# a.map() se utiliza para sustituir cada valor en una serie por otro valor, funciona exclusivamete con series o a columna de dataframe
# se le puede pasar un diccionario o una funcion para realizar mapeo
# tranforma los valores de una serie
import pandas as pd
import numpy as np
df = pd.DataFrame({'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla'],
                   'Poblacion': [3305408, 1636732, 791413, 688592]})
# Mapeo usando un diccionario
mapeo_ciudades = {'Madrid': 1, 'Barcelona': 2, 'Valencia': 3, 'Sevilla': 4}
df['Codigo_Ciudad'] = df['Ciudad'].map(mapeo_ciudades)
print(df)
```

```python
df_numerico = df[['Poblacion', 'Codigo_Ciudad']]

def formato_miles(valor):
  return f'{valor: ,}'

df_formateado = df_numerico.map(formato_miles)

df_formateado
```

```python
#Apply(): Operaciones sobre filas o columnas
# El mewtodo apply() es mas versatil y puede operar tanto en DataFrame como en series. Cuando se usa en un dataframe puede aplicar una funcion a lo largode un eje (filas o columna)

# Aplicar una funcion a auna columna serie
def categorizar_poblacion(poblacion):
  if poblacion > 2000000:
    return 'Muy grande'
  elif poblacion > 1000000:
    return 'Grande'
  else:
    return 'Mediana'
df['Categoria_Poblacion'] = df['Poblacion'].apply(categorizar_poblacion)
df
```

```python
# Aplicar una funcion a travez de las filas (axis=1)
def info_ciudad(fila):
  return f'{fila["Ciudad"]} tiene el codigo {fila["Codigo_Ciudad"]}'
df['Info_Ciudad'] = df.apply(info_ciudad, axis=1)
df
```
