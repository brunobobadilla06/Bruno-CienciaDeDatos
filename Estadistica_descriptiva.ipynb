```python
# Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid') # Establece un tema visual agradable para los graficos de Seaborn

# Crear un diccionario con los datos de los estudiantes
datos = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Sofia', 'Elena', 'Diego', 'Laura'],
    'Edad': [20, 22, 19, 21, 23, 20, 22, 19, 21, 23],
    'Calificacion': [85, 92, 78, 88, 95, 80, 90, 75, 88, 90],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona']
}

# Crear un DataFrame de pandas a partir del dicicionario
df_estudiantes = pd.DataFrame(datos)

# Mostrar el DataFrame
df_estudiantes
```

```python
# Media con la columna calificacion
media = df_estudiantes['Calificacion'].mean()
print('Media:', media)

# Media de la edad
media_edad = df_estudiantes['Edad'].mean()
print('Media de la edad:', media_edad)
```

```python
# Mediana con la culumna calificacion
mediana = df_estudiantes['Calificacion'].median()
print('Mediana:', mediana)

# Mediana de la edad
mediana_edad = df_estudiantes['Edad'].median()
print('Mediana de la edad:', mediana_edad)
```

```python
# Moda de la calificacion
moda_calificacion = df_estudiantes['Calificacion'].mode()
print('Moda de la calificacion:', moda_calificacion)

# Moda de la edad
moda_edad = df_estudiantes['Edad'].mode()
print('Moda de la edad:', moda_edad)

# Tambien podemos calcular la moda para datos categoricas
moda_ciudad = df_estudiantes['Ciudad'].mode()
print('Moda de la ciudad:', moda_ciudad)
```

```python
# Medidas de dispersion
# Rango
# Calcular el minimo y maximo de la calificacion
min_calificacion = df_estudiantes['Calificacion'].min()
max_calificacion = df_estudiantes['Calificacion'].max()
rango_calificacion = max_calificacion - min_calificacion
print(f'Calificacion Minima: {min_calificacion}')
print(f'Calificacion Maxima: {max_calificacion}')
print('Rango de la calificacion:', rango_calificacion)
```

```python
# Varianza
varianza_calificacion = df_estudiantes['Calificacion'].var()
print(f'Varianza de la calificacion: {varianza_calificacion:.2f}')
```

```python
# Desviacion estandar
desviacion_estandar_calificacion = df_estudiantes['Calificacion'].std()
print(f'Desviacion estandar de la calificacion: {desviacion_estandar_calificacion:.2f}')
# Interpretacion: En promedio, las calificaciones tienden a desviarse unos 6.49 puentos de la media 86.1
```

```python
# Cuartiles e IQR
# Podemos obtenerlos con .quartile() o directmante de .describe()
q1_cuantile = df_estudiantes['Calificacion'].quantile(0.25)
q3_cuantile = df_estudiantes['Calificacion'].quantile(0.75)
iqr = q3_cuantile - q1_cuantile

print(f'Primer cuartil (Q1): {q1_cuantile:.2f}')
print(f'Tercer cuartil (Q3): {q3_cuantile:.2f}')
print(f'Rango intercuartil (IQR): {iqr:.2f}')
```

```python
# Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid') # Establece un tema visual agradable para los graficos de Seaborn

# Crear un diccionario con los datos de los estudiantes
datos = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Sofia', 'Elena', 'Diego', 'Laura'],
    'Edad': [20, 22, 19, 21, 23, 20, 22, 19, 21, 23],
    'Calificaciones': [85, 92, 78, 88, 95, 80, 90, 75, 88, 90],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona']
}

# Crear un DataFrame de pandas a partir del dicicionario
df_estudiantes = pd.DataFrame(datos)


# Histograma de calificaciones
plt.figure(figsize=(8, 5)) # Define el tamaño de la figura
sns.histplot(data = df_estudiantes, x='Calificaciones', kde=True) # añade la curva de densidad
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.show()
# Observaciones del histograma
# Parece haber una concentracion de calificaciones alrededor de 80-87
# La distribucion no es perfectamente simetrica, quizas ligeramente sesgada a la derecha (cola mas larga)
```

```python
# Boxplot de calificaciones
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_estudiantes, y='Calificaciones')
plt.title('Boxplot de Calificaciones')
plt.xlabel('Calificaciones')
plt.show()
#
# Observaciones del boxplot
# La linea central(mediana) esta cerca de 87
# La caja de (IQR) va aproximadamente de 82 a 90
# Los bigotes muestran el rango general de los datos tipicos
# No aparecen outliers puentos fuera de los bigotes  en este dataset
```

```python
# Grafico de barras para ciudades
# Primero contemos cuantos estudiantes hay por ciudad
conteo_ciudades = df_estudiantes['Ciudad'].value_counts()
print(conteo_ciudades)

# Ahora el grafico de barras
plt.figure(figsize=(8, 5))
sns.countplot(data=df_estudiantes, x='Ciudad', order=conteo_ciudades.index, palette='pastel') # Order me ordena de mayor a menor
plt.title('Conteo de Estudiantes por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Numero de estudiantes')
plt.show()
#
# Observaciones del grafico de barras:
# Madrid es la ciudad mas frecuentes
```

Para formular TUS hipótesis:

Explora tus datos: Mira los nombres de las columnas, los tipos de datos, y calcula estadísticas descriptivas básicas (medias, medianas, conteos, etc.).

Piensa en el objetivo: ¿Qué quieres descubrir o probar? ¿Qué decisiones quieres tomar basadas en los datos?

Identifica variables clave: ¿Cuáles son las métricas importantes (ventas, clics, tiempo, errores)? ¿Cuáles son los factores que podrían influir en ellas (inversión, tipo de cliente, diseño, región)?

Formula una pregunta específica: “Me pregunto si los clientes que usan la app móvil compran más que los que usan la web.”

Conviértela en una afirmación testable (hipótesis): “Los clientes que utilizan la aplicación móvil tienen un valor promedio de compra significativamente mayor que los clientes que utilizan el sitio web.”

Ejemplo

Hipótesis Nula (H0):
Las características clínicas y demográficas disponibles (edad, sexo, tipo de dolor de pecho, etc.) no son suficientes para predecir la presencia de enfermedad cardíaca con una precisión significativamente mejor que el azar.

Hipótesis Alternativa (H1):
Es posible construir un modelo de clasificación (Random Forest) utilizando las características clínicas y demográficas que prediga la presencia de enfermedad cardíaca con una precisión y AUC (Area Under Curve) significativamente superiores al 50% (azar). Esperamos alcanzar una precisión > 75%.
