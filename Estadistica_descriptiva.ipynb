```python
# Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid') # Establece un tema visual agradable para los graficos de Seaborn

# Crear un diccionario con los datos de los estudiantes
datos = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Sofia', 'Elena', 'Diego', 'Laura'],
    'Edad': [20, 22, 19, 21, 23, 20, 22, 19, 21, 23],
    'Calificacion': [85, 92, 78, 88, 95, 80, 90, 75, 88, 90],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Barcelona']
}

# Crear un DataFrame de pandas a partir del dicicionario
df_estudiantes = pd.DataFrame(datos)

# Mostrar el DataFrame
df_estudiantes
```

```python
# Media con la columna calificacion
media = df_estudiantes['Calificacion'].mean()
print('Media:', media)

# Media de la edad
media_edad = df_estudiantes['Edad'].mean()
print('Media de la edad:', media_edad)
```

```python
# Mediana con la culumna calificacion
mediana = df_estudiantes['Calificacion'].median()
print('Mediana:', mediana)

# Mediana de la edad
mediana_edad = df_estudiantes['Edad'].median()
print('Mediana de la edad:', mediana_edad)
```

```python
# Moda de la calificacion
moda_calificacion = df_estudiantes['Calificacion'].mode()
print('Moda de la calificacion:', moda_calificacion)

# Moda de la edad
moda_edad = df_estudiantes['Edad'].mode()
print('Moda de la edad:', moda_edad)

# Tambien podemos calcular la moda para datos categoricas
moda_ciudad = df_estudiantes['Ciudad'].mode()
print('Moda de la ciudad:', moda_ciudad)
```

```python
# Medidas de dispersion
# Rango
# Calcular el minimo y maximo de la calificacion
min_calificacion = df_estudiantes['Calificacion'].min()
max_calificacion = df_estudiantes['Calificacion'].max()
rango_calificacion = max_calificacion - min_calificacion
print(f'Calificacion Minima: {min_calificacion}')
print(f'Calificacion Maxima: {max_calificacion}')
print('Rango de la calificacion:', rango_calificacion)
```

```python
# Varianza
varianza_calificacion = df_estudiantes['Calificacion'].var()
print(f'Varianza de la calificacion: {varianza_calificacion:.2f}')
```

```python
# Desviacion estandar
desviacion_estandar_calificacion = df_estudiantes['Calificacion'].std()
print(f'Desviacion estandar de la calificacion: {desviacion_estandar_calificacion:.2f}')
# Interpretacion: En promedio, las calificaciones tienden a desviarse unos 6.49 puentos de la media 86.1
```

```python
# Cuartiles e IQR
# Podemos obtenerlos con .quartile() o directmante de .describe()
q1_cuantile = df_estudiantes['Calificacion'].quantile(0.25)
q3_cuantile = df_estudiantes['Calificacion'].quantile(0.75)
iqr = q3_cuantile - q1_cuantile

print(f'Primer cuartil (Q1): {q1_cuantile:.2f}')
print(f'Tercer cuartil (Q3): {q3_cuantile:.2f}')
print(f'Rango intercuartil (IQR): {iqr:.2f}')
```

```python
# Histograma de calificaciones
plt.figure(figsize=(8, 5))
sns.histplot(data = df_estudiantes, x='Calificaciones', kde=True)
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificacion')
plt.ylabel('Frecuencia')
plt.show()
```
