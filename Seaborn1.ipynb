**Grafico Categoricos(Categorical plots)**
Comparan una variable numerica a traves de diferentes categorias

***Barplot:*** Este tipo de gráfico de barras muestra la relación entre una variable numérica y una variable categórica. Generalmente, muestra la media de la variable numérica para cada categoría, junto con barras de error que indican la variabilidad. Es útil para comparar el valor promedio de una medida entre diferentes grupos.

**Boxplot** (Diagrama de Caja): El boxplot resume la distribución de una variable numérica para cada categoría. Muestra la mediana, los cuartiles (25% y 75%), los valores atípicos y el rango de los datos. Es excelente para visualizar la dispersión y detectar posibles valores atípicos dentro de cada categoría.

**Violinplot: **Similar al boxplot, el violinplot también muestra la distribución de una variable numérica por categoría. Sin embargo, en lugar de solo mostrar cuartiles, muestra la densidad de probabilidad de los datos en diferentes valores. Es útil para ver la forma de la distribución dentro de cada categoría, especialmente si la distribución es multimodal.

**Countplot: **Este gráfico es básicamente un histograma para variables categóricas. Muestra el número de ocurrencias de cada categoría en un conjunto de datos. Es útil para visualizar la frecuencia de cada nivel en una variable categórica.

**Catplot:** Este es un envoltorio (wrapper) de alto nivel para varios tipos de gráficos categóricos. Permite crear fácilmente diferentes tipos de gráficos categóricos (como barplot, boxplot, violinplot, etc.) y dividirlos por otras variables categóricas usando facetas. Es muy flexible para explorar relaciones entre variables categóricas y numéricas.

```python
# Introduccion a Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Cargar el data set
df = pd.read_csv('https://raw.githubusercontent.com/brunobobadilla06/Bruno-CienciaDeDatos/refs/heads/main/dataset_ventas_mensuales.csv')

# Barplot: Ventas promedio por region
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='region', y='ventas', ci='sd', palette='viridis') # ci='sd' muestra la desviacio estandar
plt.title('Ventas Promedio por Region (con Desviacion Estandar )')
plt.xlabel('Region')
plt.ylabel('Ventas Promedio')
plt.show()
```

# Graficos de relacion(Relational Plots)
Exploran la relacion entre dos variables numericas

Scatterplot: Muestra la relacion entre dos variables numericas. Se puede añadir informacion categorica con hue, size, style.

Lineplot: Ideal para visualizar tendencias a lo largo del tiempo o de otravariable continua.

Relplot: Figura-nivel para scatterplots y lineplos, facilita la creacion de subgraficos basadas en categorias.

```python
# Introduccion a Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Cargar el data set
df = pd.read_csv('https://raw.githubusercontent.com/brunobobadilla06/Bruno-CienciaDeDatos/refs/heads/main/dataset_ventas_mensuales.csv')

# # Scatterplot: Relacion entre ventas y satisfaccion del cliente
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='ventas', y='satisfaccion_cliente', hue='region', style='producto', s=100) # s=100 aumenta el tamaño de los puntos
plt.title('Relacion entre Ventas y Satisfaccion del Cliente')
plt.xlabel('Ventas')
plt.ylabel('Satisfaccion del Cliente')
plt.legend(title='Leyenda', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
#
```

```python
# Lineplot: Ventas a lo largo del tiempo por producto
plt.figure(figsize=(14,7))
sns.lineplot(data=df, x='mes', y='ventas', hue='producto', style='region', marker='o')
plt.title('Ventas a lo largo del Tiempo por Producto y region')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.xticks(rotation=45) # Rotar etiquetas eje x para legibilidad
plt.legend(title='Leyenda', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Parametros importantes en seaborn:
**data:** Este es fundamental. Especifica el DataFrame que contiene los datos que quieres visualizar. Casi todas las funciones de Seaborn lo requieren.
**x y y:** Estos parámetros definen las columnas del DataFrame que se usarán para los ejes horizontal (x) y vertical (y) del gráfico. Son esenciales para determinar qué variables se van a plotear.
**hue:** Permite añadir una dimensión categórica al gráfico. Seaborn usará diferentes colores para representar las categorías de la variable especificada en hue. Esto es muy útil para ver cómo una relación o distribución varía entre diferentes grupos.
**size:** En algunos gráficos (como scatterplot), puedes usar este parámetro para variar el tamaño de los marcadores en función de los valores de otra variable (generalmente numérica).
**style:** Similar a hue y size, style te permite representar una variable categórica utilizando diferentes estilos de marcadores o líneas.
**palette:** Controla la paleta de colores utilizada en el gráfico, especialmente cuando se usa hue. Seaborn ofrece una amplia variedad de paletas predefinidas (como 'viridis', 'plasma', 'rocket', etc.), o puedes pasar una lista de colores personalizados.
**ci / errorbar:** En gráficos que muestran estimaciones (como lineplot o barplot), este parámetro controla la representación de los intervalos de confianza. Puedes especificar diferentes métodos (como 'sd' para la desviación estándar, 'ci' para el intervalo de confianza) o desactivarlos (None). En versiones más recientes de Seaborn, se prefiere errorbar sobre ci.
**kind:** Usado en funciones de alto nivel como catplot o relplot, especifica el tipo de gráfico subyacente que se debe dibujar (por ejemplo, 'bar', 'box', 'violin' para catplot; 'scatter', 'line' para relplot).
**col y row**: Utilizados en funciones de alto nivel con facetas (catplot, relplot), permiten crear subgráficos organizados en columnas (col) o filas (row) en función de las categorías de una variable.
**height y aspect:** Controlan el tamaño de la figura. height establece la altura de cada faceta (en el caso de gráficos con facetas), y aspect controla la relación de aspecto (ancho/alto) de cada faceta.
**bins:** En histogramas (histplot), define el número de contenedores (bins) o los bordes de los contenedores para agrupar los datos.
**kde:** En histogramas o gráficos de distribución, un valor booleano que indica si se debe superponer una estimación de densidad de kernel (KDE).
**fill:** En KDE plots, un valor booleano para rellenar el área bajo la curva de densidad.
**marker y linestyle:** En lineplots, permiten personalizar el estilo de los marcadores en los puntos de datos y el estilo de la línea que los conecta.
**title y etiquetas de ejes (xlabel, ylabel)**: Aunque estos son parámetros de Matplotlib, se usan comúnmente con Seaborn para añadir títulos y etiquetas a los ejes de los gráficos, mejorando su legibilidad.
**s (parámetro en Seaborn scatterplot, entre otros):** Este parámetro se utiliza para controlar el tamaño de los marcadores en un gráfico de dispersión (scatterplot). Puedes pasar un valor numérico fijo para que todos los puntos tengan el mismo tamaño, o puedes pasar el nombre de una columna del DataFrame. Si pasas el nombre de una columna, el tamaño de cada punto se escalará según el valor correspondiente en esa columna. Esto es útil para añadir una dimensión visual adicional al gráfico, representando una tercera variable numérica mediante el tamaño de los puntos. En el código que mostraste, s=100 hace que todos los puntos del scatterplot tengan un tamaño de 100.
**plt.legend():** Esta función de Matplotlib se utiliza para añadir una leyenda al gráfico. La leyenda explica qué representa cada color, estilo o tamaño en el gráfico, especialmente cuando has utilizado parámetros como hue, size o style en Seaborn. plt.legend() intenta colocar la leyenda automáticamente en una posición óptima. Los parámetros que viste en el código (title, bbox_to_anchor, loc) te permiten personalizar la leyenda:
**title: **Establece un título para la leyenda.
bbox_to_anchor: Define las coordenadas (en relación a la figura o los ejes, dependiendo del parámetro loc) donde se colocará la "caja" de la leyenda.
loc: Especifica la ubicación de la leyenda (por ejemplo, 'upper left', 'lower right', etc.) en relación a bbox_to_anchor. En el código, bbox_to_anchor=(1.05, 1) y loc='upper left' colocan la leyenda fuera del gráfico, a la derecha y alineada en la parte superior.
**plt.tight_layout():** Esta función de Matplotlib ajusta automáticamente los parámetros de la subtrama para dar espacio ajustado entre subtramas. Esto ayuda a evitar que las etiquetas de los ejes, los títulos o las leyendas se superpongan, haciendo que el gráfico sea más limpio y legible. Es una buena práctica llamarla antes de plt.show().
**plt.xticks():** Esta función de Matplotlib se utiliza para obtener o establecer las ubicaciones y etiquetas de los "ticks" (las marcas y los números o texto asociados) en el eje x. En el código que mostraste (plt.xticks(rotation=45)), se utiliza para rotar las etiquetas del eje x en 45 grados. Esto es particularmente útil cuando las etiquetas del eje x son largas (como fechas o nombres de categorías) y se superponen si se muestran horizontalmente.



# Graficos matriciales(Matrix Plots)

Visualizan datos matriciales, con matrices de correlacion



*   Heatmap: Muestra una matriz coloreando las celdas segun su valor. Util para matrices de correlacion.
*  Cluster: Similar a heatmap, pero reaodena las filas y columnas usando algoritmos de clustering para agrupar variables similares.



```python
# Introduccion a Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Cargar el data set
df = pd.read_csv('https://raw.githubusercontent.com/brunobobadilla06/Bruno-CienciaDeDatos/refs/heads/main/dataset_ventas_mensuales.csv')

# Convertir la columna 'Mes' a tipo datetime
df['mes'] = pd.to_datetime(df['mes'], format='%Y-%m-%d')

# Extraer el mes y el año (Puede ser util para agrupaciones)
df['NumeroMes'] = df['mes'].dt.month
df['Año'] = df['mes'].dt.year

# Calcular la matriz de correlacion solo columnas numericas
columnas_numericas=['ventas', 'costo_operativo', 'satisfaccion_cliente', 'NumeroMes']
matriz_correlacion = df[columnas_numericas].corr()
matriz_correlacion
#
```

```python
# Heatmap de la matriz de la correlacion
plt.figure(figsize=(10,6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Heatmap de la Matriz de Correlacion')
plt.show()
#
```
