```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Primer grafico de linea simple
# crear datos ejemplos
x = np.linspace(0, 10, 100) # 100 puntos equidistante entre 0 y 10
y = np.sin(x) # Calcular el seno de cada punto x

# Crear el grafico
plt.plot(x,y)

#AÃ±adir etiquetas y titulo
plt.xlabel('Eje X - Tiempo')
plt.ylabel('Eje Y - Amplitud')
plt.title('Mi primer grafico, Funcion seno')

# Mostrar Graficos
plt.show()
```

Anatomia de un grafico: Figura y ejes(axes)




Matplotlib:  tiene una estructura jerarquica. Lo importante es entender la diferencia entre figure y axes



*   Figure: Es el contenedor de mas alto nivel, como la ventana o la pagina completa donde todo se dibuja. Puedes pensar en ella como el lienzo en blanco.
*  Axes: Es el area dentro de la figura donde se dibujan los datos(el grafico en si, con sus ejes x e y, retiquetas). Una figura puede contener una o mas axes



Hay dos formas princuipales de usar Matplotlib




1.  INTERFAZ PYPLOT: (Estilo matlab): Como hicimos antes(plt.plot(...)). Es rapido para graficos simples, pero puede volverse confuso con graficos mas complejos o multiples graficos . Mantiene un estado global del grafico actual.
2.   Interfaz orientada a objetos(OO): Es mas aexplicita y recomendada. Creamos objetos figure y axes y luego llamamos a metodos directamente sobre esos objetos.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Crear datos de ejemplo(los mismos de antes)
x = np.linspace(0, 10, 100)
y = np.sin(x)

#Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Usar los metodos del objeto 'ax' para graficar y configurar
ax.plot(x,y)
ax.set_xlabel('Eje X - Tiempo') # Nota: set_xlabel en lugar de xlabel
ax.set_ylabel('Eje Y - Amplitud') # Nota: set_ylabel en lugar de ylabel
ax.set_title('Mi primer grafico, Funcion seno') # Nota: set_title em lugar de title

# Mostrar el grafico
plt.show()
```

Â¿Porque preferir el estilo OO?
porque es explicito. Siempre sabes sobre que objeto(ax) estas trabajando.
Esto es crucial cuando tienes multiples graficos en la misma figura.

* Acostumbrate a usar fig, ax= plt.subplots()



Tipoos de graficos:


Graficos de dispersion(Scatter Plot)
Util para ver la relacion entre dos variables numericas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Datos ejemplo (algo de ruido aleatorio)
x_scatter = np.random.rand(50) * 10 # 50 numeros aleatorios entre 0 y 10
y_scatter = x_scatter +  np.random.randn(50) * 2 # y = x + ruido aleatorio

# Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Usar los metodos del objeto 'ax' para crear grafico de dispoersion
ax.scatter(x_scatter, y_scatter)

# Etiquetas y titulo
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Grafico de dispersion')

# Mostrar el grafico
plt.show()
```

Graficos de barras(Bar chart)
Ideal para cantidades de diferentes categorias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Dtos ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Crear el grafico de barras verticales
ax.bar(categorias, valores)

# Etiquetas y titulo
ax.set_xlabel('Categorias')
ax.set_ylabel('Valores')
ax.set_title('Grafico de barras')

# Mostrar el grafico
plt.show()

# Graficos de barras  Horizontales
fig, ax = plt.subplots()
ax.barh(categorias, valores)
plt.show()
```

Histograma
Muestra la distribucion de una unica variable numerica agrupando los valores en  en bins(INTERVALOS)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# DATOS DE EJEMPLO (DISTRIBUCION NORMAL)
data = np.random.randn(1000) # 1000 valores aleatorios de una distribucion normal

# Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

#Crear el histograma
ax.hist(data, bins=30, edgecolor='black') # 30 intervalos

# Etiquetas y titulo
ax.set_xlabel('Valores')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma')

# Mostrar el grafico
plt.show()
```

Grafico circular(Pie chart)
Muestra proporciones de un todo. Usar con precaucion. Son dificiles de interpretar con muchas categorias o categorias con tamaÃ±os similares.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Datos de ejemplos
etiquetas_pie = ['Proteinas', 'Carbohidratos', 'Grasas', 'Vitaminas']
valores_pie = [30, 40, 30, 10]
colores_pie = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
explode = (0, 0.1, 0, 0) # Explorar la 2da rebanada (carbohidratos)

# Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Crear el grafico circular
ax.pie(valores_pie, explode=explode, labels=etiquetas_pie, colors=colores_pie, autopct='%1.1f%%', shadow= True, startangle=90)

# Titulo
ax.set_totle=('Grafico circular')

# Mostrar el grafico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Multiples graficos (Subplots)
# Crear Datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
data_hist = np.random.randn(500)
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Crear una figura con 2 ejes(axes)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

#axes ahora es un array de numpy 2x2 que contiene los abjetos Axes
# Accedemos a ellos como axes[fila, columna]

#subpltos 1 (ariba izquierda)
axes[0,0].plot(x,y_sin, color='blue')
axes[0,0].set_title('Funcion seno')
axes[0, 0].grid(True)

# subplots 2 (arriba derecha)
axes[0,1].plot(x,y_cos, color='red')
axes[0,1].set_title('Funcion coseno')
axes[0, 1].grid(True)

# subplots 3 (abajo izquierda)
axes[1,0].hist(data_hist, bins=30, color='green', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Histograma')

#subplots 4 (abajo derecha)
axes[1,1].bar(categorias, valores, color='orange')
axes[1,1].set_title('Grafico de barras')

# AÃ±adir un titulo general a toda la figura
fig.suptitle('Multiples graficos en una figura', fontsize=16)

# Mostrar el grafico
plt.show()
```
