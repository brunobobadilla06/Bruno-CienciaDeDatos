```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Primer grafico de linea simple
# crear datos ejemplos
x = np.linspace(0, 10, 100) # 100 puntos equidistante entre 0 y 10
y = np.sin(x) # Calcular el seno de cada punto x

# Crear el grafico
plt.plot(x,y)

#AÃ±adir etiquetas y titulo
plt.xlabel('Eje X - Tiempo')
plt.ylabel('Eje Y - Amplitud')
plt.title('Mi primer grafico, Funcion seno')

# Mostrar Graficos
plt.show()
```

Anatomia de un grafico: Figura y ejes(axes)




Matplotlib:  tiene una estructura jerarquica. Lo importante es entender la diferencia entre figure y axes



*   Figure: Es el contenedor de mas alto nivel, como la ventana o la pagina completa donde todo se dibuja. Puedes pensar en ella como el lienzo en blanco.
*  Axes: Es el area dentro de la figura donde se dibujan los datos(el grafico en si, con sus ejes x e y, retiquetas). Una figura puede contener una o mas axes



Hay dos formas princuipales de usar Matplotlib




1.  INTERFAZ PYPLOT: (Estilo matlab): Como hicimos antes(plt.plot(...)). Es rapido para graficos simples, pero puede volverse confuso con graficos mas complejos o multiples graficos . Mantiene un estado global del grafico actual.
2.   Interfaz orientada a objetos(OO): Es mas aexplicita y recomendada. Creamos objetos figure y axes y luego llamamos a metodos directamente sobre esos objetos.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Crear datos de ejemplo(los mismos de antes)
x = np.linspace(0, 10, 100)
y = np.sin(x)

#Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Usar los metodos del objeto 'ax' para graficar y configurar
ax.plot(x,y)
ax.set_xlabel('Eje X - Tiempo') # Nota: set_xlabel en lugar de xlabel
ax.set_ylabel('Eje Y - Amplitud') # Nota: set_ylabel en lugar de ylabel
ax.set_title('Mi primer grafico, Funcion seno') # Nota: set_title em lugar de title

# Mostrar el grafico
plt.show()
```

Â¿Porque preferir el estilo OO?
porque es explicito. Siempre sabes sobre que objeto(ax) estas trabajando.
Esto es crucial cuando tienes multiples graficos en la misma figura.

* Acostumbrate a usar fig, ax= plt.subplots()



Tipoos de graficos:


Graficos de dispersion(Scatter Plot)
Util para ver la relacion entre dos variables numericas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Datos ejemplo (algo de ruido aleatorio)
x_scatter = np.random.rand(50) * 10 # 50 numeros aleatorios entre 0 y 10
y_scatter = x_scatter +  np.random.randn(50) * 2 # y = x + ruido aleatorio

# Crear una figura y un conjunto de ejes(axes)
fig, ax = plt.subplots()

# Usar los metodos del objeto 'ax' para crear grafico de dispoersion
ax.scatter(x_scatter, y_scatter)

# Etiquetas y titulo
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Grafico de dispersion')

# Mostrar el grafico
plt.show()
```
